```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load old data
old_dir = 'data_comparison/old_trot_data'

test_traj={}
for filename in os.listdir(old_dir):
    f = os.path.join(old_dir, filename)
    # checking if it is a file
    if os.path.isfile(f) and filename[0] != '.':
        data_name = filename.split('.')[0]
        test_traj[data_name] = np.load(f)
        
print("Old data keys:\n--------------------")
print(test_traj.keys())

# load new data
new_dir = 'data_comparison/new_trot_data'

new_traj={}
for filename in os.listdir(new_dir):
    f = os.path.join(new_dir, filename)
    # checking if it is a file
    if os.path.isfile(f):
        data_name = filename.split('.')[0]
        try:
            new_traj[data_name] = np.load(f, allow_pickle=True)
        except:
            pass
        
print("\nNew data keys:\n--------------------")
print(new_traj.keys())
```

```python
from scipy.signal import butter, lfilter, freqz, filtfilt
import matplotlib.pyplot as plt


def butter_lowpass(cutoff, fs, order=5):
    return butter(order, cutoff, fs=fs, btype='low', analog=False)

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y


# Filter requirements.
order = 6
fs = 1000.0       # sample rate, Hz
cutoff = 10  # desired cutoff frequency of the filter, Hz
b, a = butter_lowpass(cutoff, fs, order)

data = new_base_vel[:,2]

# Get the filter coefficients so we can check its frequency response.
y = filtfilt(b,a,data)

plt.plot(new_t, data, label='data')
plt.plot(new_t, y, label='filtered')
plt.show()
```

```python
# printing each contact graph one by one
old_contact = test_traj['contact']
new_contact = new_traj['contact']

plt.plot(new_t, new_contact[:,0])
plt.show()
```

```python
def contact_cleanup(contact):
    cleaned_contact = contact.copy()
    for i in np.arange(1,contact.shape[0]):
        prev = contact[i-1]
        curr = contact[i]

        # if signal falls check if it goes up again in the next 50 indices
        for n in range(4):
            try:
                next_i = contact[i+1:i+51,n]
            except:
                next_i = contact[i+1:,n]

            if curr[n]-prev[n] < 0:
                if sum(next_i) > 0:
                    curr[n] = 1
            elif curr[n]-prev[n] > 0:
                if sum(next_i) < 10:
                    curr[n] = 0

        cleaned_contact[i] = curr
        
    return cleaned_contact
```

```python
from scipy.spatial.transform import Rotation

gen_coord = pd.read_csv('trot.csv',header=None)
com_pos = gen_coord.iloc[:,0:3]
com_quart = gen_coord.iloc[:,3:7]
joints = gen_coord.iloc[:,7:]

values = com_quart.to_numpy()
plt.plot(values[:,0])
 
def quart2eul(quat_df):
    rot = Rotation.from_quat(quat_df)
    rot_euler = rot.as_euler('xyz', degrees=False)
    euler_df = pd.DataFrame(data=rot_euler, columns=['x', 'y', 'z'])

    return euler_df

eul = quart2eul(values)
eul = eul.to_numpy()
plt.figure()
plt.plot(eul[:,0])
```
